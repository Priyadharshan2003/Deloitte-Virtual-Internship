```python
df = pd.read_csv("Task 5 Equality Table.csv")
```

```python
df
```

Create a 4th column (Equality class), classifying the equality score in those 3 types:

    Fair (+-10)
    Unfair (<-10 AND >10)
    Highly Discriminative (<-20 AND >20)


```python
# create a list of our conditions
conditions = [
    (df['Equality Score'] >= -10) & (df['Equality Score'] <= 10),
    (df['Equality Score'] > 10) & (df['Equality Score'] <= 20),
    (df['Equality Score'] < -10) & (df['Equality Score'] >= -20),
    (df['Equality Score'] > 20),
    (df['Equality Score'] < -20)
    ]
```

```python
# create a list of the values we want to assign for each condition
values = ['Fair', 'Unfair', 'Unfair', 'Highly Discriminative', 'Highly Discriminative']
```

```python
# create a new column and use np.select to assign values to it using our lists as arguments
df['Equality class'] = np.select(conditions, values)
```
